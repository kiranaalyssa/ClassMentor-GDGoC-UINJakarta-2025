## 5. Handling Missing Value
Referensi yang dapat dibaca:


*   https://ngodingdata.com/python-pandas-cara-menangani-missing-value/
*   https://www.geeksforgeeks.org/ml-handling-missing-values/



```python
# 1. Dengan menghilangkan Missing Value
# Membuat DataFrame baru yang berisi data setelah baris dengan missing value dihapus.
df_cleaned = df.dropna()

# Menampilkan DataFrame yang sudah bersih dari missing value.
print("DataFrame setelah Missing Value dihapus:")
df_cleaned

# 2. Mean, Median, and Mode Imputation (Mengganti Missing Value dengan nilai Mean, Median, atau Mode)
# Mean: df['mass'].fillna(df['mass'].mean(), inplace=True)
# Median: df['mass'].fillna(df['mass'].median(), inplace=True)
# Mode: df['category'].fillna(df['category'].mode()[0], inplace=True)
```

Dataset Planets yang awalnya memiliki **1035** baris berkurang menjadi **498** baris setelah menghapus data dengan nilai yang hilang (Missing Value). Perlu dicatat, metode ini kurang direkomendasikan karena dapat mengurangi banyak informasi berharga dan memperkecil ukuran dataset secara signifikan.

```python
# Periksa kembali dataset setelah menghapus nilai yang hilang.
df_cleaned.isnull().sum()
```
